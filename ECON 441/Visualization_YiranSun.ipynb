```python
import pandas as pd
import numpy as np
import ipywidgets as widget
import matplotlib.pyplot as plt
import re
```

```python
data = pd.read_csv("MoviesOnStreamingPlatforms.csv")

data_copy =data.copy() 
# clean the data first
data['IMDb'] = data['IMDb'].astype(float)
```

```python
data_copy =data.copy() 
data_copy= data_copy.dropna(axis=0,subset=['Rotten Tomatoes'])

#data_copy= data_copy.drop(labels = data_copy[data_copy['Rotten Tomatoes'].isnull()]
#                          .index, axis=0)

data_copy['Rotten Tomatoes'] = data_copy['Rotten Tomatoes'].astype(str)

pattern = re.compile(r"\%") 
data_copy['Rotten Tomatoes'] = [pattern.sub("", match) 
                                for match in data_copy['Rotten Tomatoes']]

data_copy['Rotten Tomatoes'] = data_copy['Rotten Tomatoes'].astype(int)
```
